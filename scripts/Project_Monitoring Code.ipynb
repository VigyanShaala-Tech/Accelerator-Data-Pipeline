```python
# Count the number of columns with names starting with "Project_Details"
project_details_columns = [col for col in output_data.columns if col.startswith('Project Details')]
project_details_column_count = len(project_details_columns)

print("Number of columns starting with 'Project Details':", project_details_column_count)
```

```python
#create dataframe contining only columns starting with Project Details 
project_details_df = output_data[project_details_columns]
```

```python
#Get the Project Details column containing the filled values from the latest details column. 
project_details_df['Project Details'] = project_details_df.ffill(axis=1).iloc[:, -1]


```

```python
#Convert Latest_filled_column to dataframe
Latest_filled_column=pd.DataFrame(project_details_df['Project Details'])
```

```python
#Create dataframe of the first 5 columns present in the main output_data 
output_d= output_data.iloc[:, :5]
```

```python
#Create dataframe of the last 6 columns present in the main output_data 
output_d1= output_data.iloc[:, -6:]
```

```python
#Concatenate three dataframes
output= pd.concat([output_d, Latest_filled_column, output_d1], axis=1)
```
